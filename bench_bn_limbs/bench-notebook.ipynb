# Benchmark bn em função do numero de limbs

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
```

```python
u_df = pd.read_csv('unprotected.csv')
u_df
```

```python
p_df = pd.read_csv('protected.csv')
p_df
```

```python
df = u_df.merge(p_df, on=['Limbs','Function'])
df
```

## BN_Addn

```python
df_add = df[df['Function'] == 'bn_addn']
ax = df_add.plot(x='Limbs', y=['Unprotected', 'Protected'])
ax.set_ylabel('Clock Cyles')
ax.set_title('bn_addn performance comparison')
plt.show()
```
